```python
import pdfplumber
```

```python
path_to_pdf ='1st 2022.pdf'
pdf = pdfplumber.open(path_to_pdf)
```

```python
first_page = pdf.pages[0]
print(first_page)
```

```python
text = first_page.extract_text()
print(text)
```

```python
pdf.pages
```

```python
for pg in pdf.pages:
   text = pg.extract_text()
   print(text)
```

```python
tables = first_page.extract_tables()
for table in tables:
    for row in table:
        print(row)
```
